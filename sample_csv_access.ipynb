## CSVファイルの読み込み
特定の列を抽出する方法についてまとめたもの。<br>
多次元配列の特定列を抽出する方法が分からなかったため。

### ポイント
**スライス`[:,0]`を使用して、多次元配列のうち、特定の列を抽出する**

### 参考サイト
1. https://note.nkmk.me/python-numpy-loadtxt-genfromtxt-savetxt/
1. https://note.nkmk.me/python-slice-usage/



```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ------------------------
# CSVファイル 読み込み
# ------------------------

# NOTE: encodingで文字コードを指定しないと下記エラーが出た。
# 'cp932' codec can't decode byte 0xef in position 0: illegal multibyte sequence
#NOTE: skiprows で省略する行を指定できる。
data =np.loadtxt( fname="test_sin.csv",
            dtype="float",       
            delimiter=",",
            skiprows= 1,
            encoding='utf8'
    )
# エクセルで正弦波を作る方法
# https://jugo-blog.com/trigonometry-supplement
```

```python
# ------------------------
# 設定値
# ------------------------

# 正弦波
#NOTE: スライス[:,0]を使用して、多次元配列のうち、特定の列を抽出する
sin = data[:,1]

# データサイズ
# CSVのデータサイズを変数に代入
data_size = data[:,0].size
print('data size is % d' % data_size)

# サンプリング周波数
fs = 1 * 10**6

# 時間
t = np.arange( 1/ fs, data_size / fs, 1 / fs )


# ------------------------
# グラフの表示
# ------------------------

# グラフ枠の作成
fig = plt.figure()
ax1 = plt.subplot(111)

# ---- 時間信号のデータ表示 ---- #

ax1.plot( t , sin )

# ラベル
ax1.set_xlabel("Time [s]")
ax1.set_ylabel("Signal")
```
